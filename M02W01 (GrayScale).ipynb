```python
#Question 12
import matplotlib . image as mpimg
img = mpimg.imread ('dog.jpeg')
gray_img_01 = (np.max(img, axis = 2) + np.min(img, axis = 2)) / 2
plt.imshow(gray_img_01, cmap = 'gray')
plt.axis("off")
gray_img_01[0, 0]
```

```python
#Question 13
img = mpimg.imread ('dog.jpeg')
gray_img_02 = np.sum(img , axis = 2) / 3
plt.imshow(gray_img_02, cmap = 'gray')
plt.axis("off")
gray_img_02[0 , 0]
```

```python
#Question 14
img = mpimg.imread ('dog.jpeg')
gray_img_03 = img[:,:,0] * 0.21 + img[:,:,1] * 0.72 + img[:,:,2] * 0.07
plt.imshow(gray_img_03, cmap = 'gray')
plt.axis("off")
gray_img_03[0 , 0]
```
